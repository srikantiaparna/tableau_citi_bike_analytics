```python
### JC- CitiBikeTripData
import os, shutil
import pandas as pd
import glob

#import csv files from folder
path = r'C:/Users/aparn/Tableau_Data/Resources'
allFiles = glob.glob(path + "/*.csv")
with open('C:/Users/aparn/Tableau_Data/Output/output.csv', 'wb') as outfile:
    #C:\Users\aparn\Tableau_Data\Resources
    for i, fname in enumerate(allFiles):
        with open(fname, 'rb') as infile:
            if i != 0:
                infile.readline()  # Throw away header on all but first file
# Block copy rest of file from input to output without parsing
            shutil.copyfileobj(infile, outfile)
            print(fname + " has been imported.")
```

```python
# making data frame from csv file 
data = pd.read_csv("/Users/aparn/Tableau_Data/Output/output.csv") 
# Preview the first 5 lines of the loaded data 
data.head()
```

```python
# making new data frame with dropped NA values 
new_data = data.dropna(axis = 0, how ='any') 
  
# comparing sizes of the data frames 
print("Old data frame length:", len(data), "\nNew data frame length:",  
       len(new_data), "\nNumber of rows with at least 1 NA value: ", 
       (len(data)-len(new_data)))
```

```python
# Export to csv
new_data.to_csv("C:/Users/aparn/Tableau_Data/CleanData/citibikejc_tripdata.csv")
```
